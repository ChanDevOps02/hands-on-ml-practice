## 7. 하이퍼파라미터 세부조정
모델을 선택했다면 모델을 본격적으로 학습시키기전에 모델의 하이퍼파라미터를 세부조정을 하여 모델의 성능을 향상시킬 수 있다. 이때 sklearn.model_selection의 GridSearchGV를 사용하면 모델을 학습시키는데 있어 가장 모델의 성능을 향상시킬 수 있는 하이퍼파라미터를 알아낼 수 있다.

```python
from sklearn.model_selection import GridSearchCV

param_grid = [{'n_estimators' : [3, 10, 30], 'max_features' : [2, 4, 6, 8]},
                         {'bootstrap' : [False], 'n_estimators' : [3, 10], 'max_features' : [2, 3, 4]}]
forest_reg = RandomForestRegressor()
grid_search = GridSearchCV(forest_reg, param_grid, cv = 5, scoring = 'neg_mean_squared_error', return_train_score = True)
grid_search.fit(housing_prepared, housing_label)
```

```python
grid_search.best_estimator_
```

> ℹ️ **부가 설명 | 하이퍼파라미터의 쌍에 대한 각각의 RMSE score확인하기**
cv_results_객체에 각 하이퍼파라미터 쌍에 대한 MSE score가 저장되어 있으며 이를 확인할 수 있다

```python
cvres = grid_search.cv_results_
for mean_score, params in zip(cvres['mean_test_score'], cvres['params']):
    print(np.sqrt(-mean_score), params)
```
